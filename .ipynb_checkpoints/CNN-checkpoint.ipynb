```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage.color import rgb2gray
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import OneHotEncoder as OHE
from sklearn.model_selection import train_test_split
```

## Reference: http://www.cs.toronto.edu/~kriz/cifar.html

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
file_name_list = ['data_batch_1','data_batch_2','data_batch_3','data_batch_4','data_batch_5','test_batch']

birds_data = np.empty((0,3072), np.uint8())
birds_label = list()

for file in file_name_list:
    file_path = f'cifar-10-python/cifar-10-batches-py/{file}'
    file_dict = unpickle(file_path)
    
    labels = file_dict[list(file_dict.keys())[1]] #index 1 is the 0-9 lables of each one of 10000 imgs in this batch
    data = file_dict[list(file_dict.keys())[2]] #index 2 is the 10000*3072 np array of each imgs
    image_name = file_dict[list(file_dict.keys())[3]] #index 3 is the image name of each of the 10000 imgs
    
    bird_index_list = [] #list that holds all index of birds image for each data batch
    for index, label in enumerate(labels):
        if label==2: # 2 is the label for bird
            bird_index_list.append(index)
            birds_label.append(index) #list that holds all index of birds image for all data batch
            
    birds_data_temp = data[bird_index_list] #np array that holds data for all birds for each data batch
    birds_data = np.append(birds_data, birds_data_temp, axis=0) #np array that holds data for all birds for all data batch

'''
The first 1024 entries contain the red channel values, the next 1024 the green, 
and the final 1024 the blue. The image is stored in row-major order, 
so that the first 32 entries of the array are the red channel values of the first row of the image.

So we need to convert [[1,2,3],[4,5,6],[7,8,9]] to [[1,4,7],[2,5,8],[3,6,9]]
'''
birds_data_reshaped_list = [[[birds_data[j][i], birds_data[j][i+1024], birds_data[j][i+2048]] for i in range(int(len(birds_data[j])/3))] for j in range(len(birds_data))]
birds_data_reshaped_arr = np.array(birds_data_reshaped_list)
birds_data_reshaped = birds_data_reshaped_arr.reshape(-1,32,32,3)

print(f'Shape of birds_data before reshaping is {birds_data.shape}')
print(f'Shape of birds_data after reshaping is {birds_data_reshaped.shape}')
print(f'Length of birds_label is {len(birds_label)}')
```

```python
#original img
plt.imshow(birds_data_reshaped[511])
```

```python
#k-means need the input data to be 2-d array, so input will be converted to shape (-1,3)
#input data will be each pixels and each pixels is represented by rgb
birds_data_kmeans_input = birds_data_reshaped.reshape(-1, 3)
kmeans = KMeans(n_clusters=4, random_state=1218, n_init=10).fit(birds_data_kmeans_input)
kmeans_colors = kmeans.cluster_centers_
pred_label = kmeans.labels_

#use predicted label and predicted color to re-color the img
birds_data_kmeans_reshaped = np.array([kmeans_colors[pred] for pred in pred_label], dtype=np.uint8()).reshape(-1, 32, 32, 3)

#print out the img converted to colors from kmeans
plt.imshow(birds_data_kmeans_reshaped[511])
```

```python
#suggested colors instead of kmeans colors
suggested_colors = np.array([
    [0,   0, 128], #Navy
    [230, 25, 75], #Red
    [170,255,195], #Mint
    [255,255,255] #White
], dtype=np.uint8())

birds_data_suggested_reshaped = np.array([suggested_colors[pred] for pred in pred_label], dtype=np.uint8()).reshape(-1, 32, 32, 3)
plt.imshow(birds_data_suggested_reshaped[511])
```

## Reference
- https://scikit-image.org/docs/dev/api/skimage.color.html#skimage.color.rgb2gray
- https://stackoverflow.com/questions/3823752/display-image-as-grayscale-using-matplotlib

```python
#convert the img to gray scale
birds_data_gray = np.array([rgb2gray(birds_data_reshaped[i]) for i in range(len(birds_data_reshaped))])
plt.imshow(birds_data_gray[511], cmap='gray')

birds_data_gray[0].shape
```

## Reference
- https://towardsdatascience.com/simple-introduction-to-convolutional-neural-networks-cdf8d3077bac
- https://missinglink.ai/guides/convolutional-neural-networks/python-convolutional-neural-network-creating-cnn-keras-tensorflow-plain-python/
- https://adventuresinmachinelearning.com/keras-tutorial-cnn-11-lines/
- https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2D

```python
cnn = keras.models.Sequential()

cnn.add(keras.layers.Conv2D(filters=64, kernel_size=(5, 5), strides=(1, 1), input_shape=(32,32,1), padding='same'))
cnn.add(keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1, 1), padding='same'))
cnn.add(keras.layers.Conv2D(filters=64, kernel_size=(5, 5), strides=(1, 1), padding='same'))
cnn.add(keras.layers.MaxPooling2D(pool_size=(2, 2),strides=(1, 1), padding='same'))
cnn.add(keras.layers.Dense(4, activation="softmax"))

cnn.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=["accuracy"])
cnn.summary()
```

## Reference
- https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html
- https://stackoverflow.com/questions/47790854/how-to-perform-onehotencoding-in-sklearn-getting-value-error
- https://machinelearningmastery.com/display-deep-learning-model-training-history-in-keras/

```python
encoder = OHE(handle_unknown='ignore')
pred_reshaped = encoder.fit_transform(pred_label.reshape(-1,1)).toarray().reshape(-1,32,32,4)
birds_data_gray_reshaped = birds_data_gray.reshape(-1,32,32,1)

#use first 5400 images as training set and rest 600 as test set
x_train = birds_data_gray_reshaped[:5400]
x_test = birds_data_gray_reshaped[5400:]
y_train = pred_reshaped[:5400]
y_test = pred_reshaped[5400:]

#train the model
history = cnn.fit(x_train, y_train, epochs=30, validation_data=(x_test, y_test))
```

```python
# summarize history for error
plt.plot(history.history['loss'], label='Training error')
plt.plot(history.history['val_loss'], label='Test error')
plt.title('model accuracy')
plt.ylabel('Training and test errors')
plt.xlabel('epoch')
plt.legend(loc='upper right')
plt.show()
```

```python
pred_color = cnn.predict(x_test[:10]) #first ten pic in test set
pred_color_reshape = pred_color.reshape(-1,4) # reshape to -1 4 to get the color
output_img = np.array([kmeans_colors[np.argmax(pixel)] for pixel in pred_color_reshape], dtype=np.uint8()).reshape(-1,32,32,3)

fig, axes = plt.subplots(10, 4, figsize=(10, 23))
ax = axes.ravel()

for i in range(10):
    ax[4*i].imshow(birds_data_reshaped[5400 + i]) #original img
    ax[4*i+1].imshow(birds_data_kmeans_reshaped[5400 + i]) #kmeans colorized img
    ax[4*i+2].imshow(x_test[i].reshape(32,32), cmap='gray') #gary scale img, input of CNN
    ax[4*i+3].imshow(output_img[i]) #CNN colorized img

ax[0].set_title('Original')
ax[1].set_title('Kmeans')
ax[2].set_title('CNN input/grayscale')
ax[3].set_title('CNN output/Colored')

fig.tight_layout()
plt.show()
```
